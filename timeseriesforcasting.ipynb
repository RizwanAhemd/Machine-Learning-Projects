```python
import pandas as pd
```

```python
import yfinance as yf
```

```python
import matplotlib.pyplot as plt
```

```python
ticker_symbol = "AMZN"

# Create a Ticker object
ticker = yf.Ticker(ticker_symbol)

# Fetch historical market data
data = ticker.history(period="3y")  # data for the last year
print("Historical Data:")
print(data)
```

```python
data.sample(5)
```

```python
data.info()
```

```python
stock_data=pd.DataFrame(data=data,columns=data.columns)
```

```python
stock_data['Date']=pd.to_datetime(stock_data.iloc[:,0])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(stock_data.index, stock_data['Close'], label='Close over time')
plt.xlabel("Date")
plt.ylabel("Close")
plt.title("Visualization of Close")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculate 7-day moving average
stock_data['Seven_days_ma'] = stock_data['Close'].rolling(window=7).mean()
stock_data.Seven_days_ma.isna().sum()
stock_data.isna().sum()
stock_data.dropna(inplace=True)
```

```python
stock_data.drop('7_day_MA',axis=1,inplace=True)
```

```python
stock_data.sample(6)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(stock_data.index, stock_data['Close'], label='Original Close', color='blue')
plt.plot(stock_data.index, stock_data['Seven_days_ma'], label='Seven_days_ma', color='orange')
plt.title('Close Price vs. 7-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
