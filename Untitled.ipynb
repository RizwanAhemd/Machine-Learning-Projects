```python
import pandas as pd
data = pd.read_csv('loan.csv')
data=data.drop('Loan_ID',axis=1)
data.head()
```

```python
data.isna().sum()
```

```python
data.fillna({'Gender':data.Gender.mode()[0],'Self_Employed':data.Self_Employed.mode()[0],'Dependents':data.Dependents.mode()[0],'Loan_Amount_Term':data.Loan_Amount_Term.mode()[0],'Credit_History':data.Credit_History.mode()[0]},inplace=True)
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data.Property_Area=data.Property_Area.apply(lambda x:1 if x=='Rural' else 0)
```

```python
data.Dependents.replace(['1', '0', '2', '3+'],[1,0,2,3],inplace=True)
```

```python
ohe = OneHotEncoder()
data.head(1)
```

```python
enc_data=ohe.fit_transform(data[['Married','Gender','Self_Employed','Education']]).toarray()
```

```python
enc=pd.DataFrame(enc_data,columns=ohe.get_feature_names_out())
tranform_data=pd.concat([data,enc],axis=1)
```

```python
tranform_data.drop(['Gender','Married','Education','Self_Employed'],axis=1,inplace=True)
```

```python
tranform_data.Loan_Status=tranform_data.Loan_Status.apply(lambda x:1 if x=="Y" else 0)
```

```python
tranform_data.head()
```

```python
X=tranform_data.drop('Loan_Status',axis=1)
y=tranform_data.Loan_Status
```

```python
X.head(1)
y.head(1)
y.value_counts()
```

```python
# from imblearn.over_sampling import SMOTE
# smote=SMOTE(sampling_strategy='minority') 
# X,y=smote.fit_resample(X,y)
# y.value_counts()
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE(sampling_strategy='minority')
```

```python
X,y=smote.fit_resample(X,y)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=123)
model= LogisticRegression()
```

```python
X_train.shape , y_train.shape, y_test.shape ,X_test.shape
```

```python
model.fit(X_train,y_train)
```

```python
y_pred_train=model.predict(X_train)
```

```python
y_pred_test=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_train,y_pred_train)
```

```python
confusion_matrix(y_test,y_pred_test)
```

```python
print(classification_report(y_train,y_pred_train))
```

```python
print(classification_report(y_test,y_pred_test))
```
