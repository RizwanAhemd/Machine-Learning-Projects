```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df= pd.read_csv('Real estate.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.corr()
```

```python
'No', 'X1 transaction date''X5 latitude', 'X6 longitude'
X=df.drop(['No', 'X1 transaction date','X5 latitude', 'X6 longitude'],axis=1)
```

```python
X.head()
```

```python
X.columns=['house_age','near_mrt_distance','number_of_store','price']
```

```python
X.head()
```

```python
y=X.price
```

```python
X=X.drop('price',axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python

from sklearn.model_selection import train_test_split
```

```python
# Splitting data into train test 
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=123,test_size=.30)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
y_test.shape
```

```python



from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#Checking model performance
```

```python
y_train_pred=model.predict(X_train)
```

```python
y_test_pred=model.predict(X_test)
```

```python
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```

```python
mean_squared_error(y_train,y_train_pred)
```
