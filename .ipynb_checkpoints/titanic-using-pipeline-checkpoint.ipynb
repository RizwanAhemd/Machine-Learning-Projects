```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# Let's Plan

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']), df['Survived'],test_size=0.2, random_state=42)
```

```python
# imputation transformer
trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])],remainder='passthrough')
```

```python
# one hot encoding
trf2 = ColumnTransformer([
    ('ohe_sex',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1]),
    ('ohe_embarked',OneHotEncoder(sparse_output=False, handle_unknown='ignore'),[6])],remainder='passthrough')
```

```python
# Scaling
trf3 = ColumnTransformer([('scale',MinMaxScaler(),slice(0,10))])
```

```python
# train the model
trf5 = DecisionTreeClassifier()
```

```python
trf6=[1,2,3,4,5,None]
```

# Create Pipeline

# Pipeline Vs make_pipeline

Pipeline requires naming of steps, make_pipeline does not.

(Same applies to ColumnTransformer vs make_column_transformer)

```python
# Alternate Syntax
pipe = make_pipeline(trf1,trf2,trf3,trf5,trf6)
```

```python
# train
pipe.fit(X_train,y_train)
```

# Explore the Pipeline

```python
# Code here
pipe.named_steps
```

```python
# Predict
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Cross Validation using Pipeline

```python
# cross validation using cross_val_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy').mean()
```

# GridSearch using Pipeline

```python
# gridsearchcv
params = {
    'trf5__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe,param_grid=params, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

# Exporting the Pipeline

```python
# export 
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```
