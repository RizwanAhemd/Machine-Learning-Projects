```python
import pandas as pd
```

```python
data  = pd.read_csv("diabetes.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
data.Outcome.unique()
```

```python
X = data.drop('Outcome', axis=1)
y = data.Outcome
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20, random_state=123)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
```

```python
y_pred_train = log_reg.predict(X_train)
```

```python
y_pred_train
```

```python
count = 0
for o,p in zip(y_train, y_pred_train):
    if o==p:
        count+=1
accuracy = f"{(count/len(X_train)) *100}%" 
accuracy
```

```python
list(zip(y_train, y_pred_train))
```

```python
y_train.shape
```

```python
y_pred_test = log_reg.predict(X_test)
```

```python
count = 0
for o,p in zip(y_test, y_pred_test):
    if o==p:
        count+=1
accuracy = f"{(count/len(X_test)) *100}%" 
accuracy
```

```python
class_0 = data[data.Outcome==0].sample(268)
```

```python
data2 = pd.concat([class_0,class_1]).sample(536)
```

```python
data2.Outcome.value_counts()
```

```python
X = data2.drop('Outcome', axis=1)
y = data2.Outcome
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20, random_state=123)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
```

```python
y_pred_train = log_reg.predict(X_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train, y_pred_train))
```

```python
y_pred_test = log_reg.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_test))
```
