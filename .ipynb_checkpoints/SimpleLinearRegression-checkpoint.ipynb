```python
import pandas as pd
data  = pd.read_csv('weight-height.csv')
```

```python
data.head()
```

```python
data.Gender.unique()
```

```python
data.info()
```

```python
data.Gender = data.Gender.apply(lambda x: 1 if x =="Male" else 0 )
```

```python
data.Gender.unique()
```

```python
data.corr()
```

```python
data.plot(kind='scatter', x='Height', y='Weight', title='Weight and Height in adults')
```

# Since the target column is Continuous value : Regression  

```python
X = data.drop(["Weight",'Gender'], axis=1)
y = data['Weight']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=0)
```

```python
X_train.shape , X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
```

```python
slr.fit(X_train,y_train)
```

```python
slr.coef_
```

```python
slr.intercept_
```

# Prediction on training Data

```python
y_pred = slr.predict(X_train)
```

# Performance Evluation
    - r2Score

```python
from sklearn.metrics import r2_score, root_mean_squared_error, mean_absolute_error
```

```python
root_mean_squared_error(y_train,y_pred)
```

```python
mean_absolute_error(y_train,y_pred)
```

```python
r2_score(y_train,y_pred)
```

```python
y_pred_test = slr.predict(X_test)
```

```python
r2_score(y_test,y_pred_test)
```

```python
def predict_weight(height):
    return slr.predict(height)*0.45
```

```python
height_inch = (float(input("Please enter your height: ")))*12
predict_weight([[height_inch]])
```
