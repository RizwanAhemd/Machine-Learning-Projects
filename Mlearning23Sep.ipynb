```python
import pandas as pd
data=pd.read_csv('weight-height.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.Gender=data.Gender.apply(lambda x:1 if x=="Male" else 0)
```

```python
data.head()
```

```python
data.corr()
```

```python
data.plot(kind='scatter',x='Height',y='Weight',title='Wieght and Height')
```

```python
X=data.drop(["Weight","Gender"],axis=1)
```

```python
y=data['Weight']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=1234)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
slr=LinearRegression()
```

```python
slr.coef_
```

```python
slr.intercept_
```

# Prediction on Training Data


```python
y_pred=slr.predict(X_train)
y_pred.shape
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error,mean_absolute_error
```

```python
root_mean_squared_error(y_train,y_pred)
```
